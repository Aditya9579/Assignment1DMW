```python
import pandas as pd
```

```python
df=pd.read_csv("movie_metadata.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
# Data Reduction
# Remove/reduce unnecessary attributes from dataframe
subset = df[['movie_title','num_user_for_reviews','genres','imdb_score','budget','gross','duration','title_year','language','actor_1_name']]
subset.head()
```

```python
# Data Cleaning
# Find missing values
subset.isnull().sum()
```

```python
# Define function for median
def impute_median(series):
    return series.fillna(series.median())
```

```python
# Fill missing value of integer column with median
subset.title_year=df['title_year'].transform(impute_median)
subset.budget=df['budget'].transform(impute_median)
subset.gross=df['gross'].transform(impute_median)
subset.duration=df['duration'].transform(impute_median)
```

```python
subset.isnull().sum()
```

```python
# Fill missing value of integer column with mode
subset['language'] = subset['language'].fillna(subset['language'].mode()[0])
subset['actor_1_name'] = subset['actor_1_name'].fillna(subset['actor_1_name'].mode()[0])
```

```python
subset.isnull().sum()
```

```python
# Read another dataset with atleast one same attribute as previous dataset
df2 = pd.read_csv('main_data.csv')
```

```python
df2.head()
```

```python
df2.info()
```

```python
# Data Reduction 
subset1 = df2[['movie_title','director_name','actor_2_name','actor_3_name']]
subset1.head()
```

```python
display(subset,subset1)
```

```python
# Data Integration
# Integrate/merge two dataframe
subset2 = pd.merge(subset,subset1,left_index=True, right_index=True)
```
